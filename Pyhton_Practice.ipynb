```python
"""Develop a Python program using the Matplotlib
library to generate a line graph where the data for
both the X-axis (Study Hours) and Y-axis (Marks
Obtained) is provided by the user.The program
must:
a) Accept a single integer N, representing the
number of data points.
b) Accept N integer values for the X-axis (Study
Hours).
c) Accept N integer values for the Y-axis (Marks
Obtained).
d) Generate a line plot with the following
customizations:
● Title: "Student Performance Analysis".
● Axis Labels: X-axis as "Hours Studied" and Y-axis as
"Marks (0-100)".
● CustomizaƟon: Use a blue color, a solid line style,
and a square marker.
● Legend: "Performance Data"."""
import matplotlib.pyplot as plt
# a) Accept a single integer N (number of data points)
n = int(input("Enter the number of data points (N): "))
 # Lists to store the axis data
study_hours = []
marks_obtained = []

# b) Accept N integer values for the X-axis (Study Hours)print(f"\nEnter {n} values for Study Hours (X-axis):")
for i in range(n):
   val = int(input(f"  Value {i+1}: "))
   study_hours.append(val)

 # c) Accept N integer values for the Y-axis (Marks Obtained)
print(f"\nEnter {n} values for Marks Obtained (Y-axis):")
for i in range(n):
  val = int(input(f"  Value {i+1}: "))
  marks_obtained.append(val)

  # d) Generate the line plot with specific customizations
  # color='blue', linestyle='-' (solid), marker='s' (square)
plt.plot(study_hours, marks_obtained, color='blue',
                 linestyle='-', marker='s', label='Performance Data')
 # Set Title
plt.title("Student Performance Analysis")

# Set Axis Labels
plt.xlabel("Hours Studied")
plt.ylabel("Marks (0-100)")

 # Add Legend
plt.legend()

  # Display the plot
plt.show()
```
